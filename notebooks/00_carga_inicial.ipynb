```python
import matplotlib.pyplot as plt
import seaborn as sns

#Estilo
sns.set_style('whitegrid')

#Histograma de Average_order_value
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.histplot(df['Average_Order_Value'], bins=20, kde=True)
plt.title('Distribución del Valor Promedio del Pedido')

#Histograma de Purchase_Frequency
plt.subplot(1,2,2)
sns.histplot(df['Purchase_Frequency'], bins=20, kde=True)
plt.title('Distribución de Frecuencia de Compra')
plt.tight_layout()
plt.show()
```

```python
#Diagrama de dispersión
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Average_Order_Value', y='Purchase_Frequency', alpha=0.5)
plt.title('Relación entre Valor Promedio y Frecuencia de Compra')
plt.xlabel('Valor Promedio del Pedido')
plt.ylabel('Frecuencia de Compra')
plt.show()
```

```python
#Escalar datos
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

X = df[['Average_Order_Value', 'Purchase_Frequency']]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
inercias = []
K_range = range(1, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inercias.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(K_range, inercias, 'bo-')
plt.xlabel('Número de clusters (k)')
plt.ylabel('Inercia')
plt.title('Método del Codo para encontrar k óptimo')
plt.grid(True)
plt.show()
```

```python
k_optimo = 4  # cámbialo si el codo indica otro valor
kmeans = KMeans(n_clusters=k_optimo, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# Ver clusters
df['Cluster'].value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='Average_Order_Value', y='Purchase_Frequency', 
                hue='Cluster', palette='viridis', alpha=0.6)
plt.title('Segmentación de clientes en 4 clusters')
plt.xlabel('Valor Promedio del Pedido')
plt.ylabel('Frecuencia de Compra')
plt.legend(title='Cluster')
plt.show()
```

```python
# Calcular promedios por cluster
perfil = df.groupby('Cluster')[['Average_Order_Value', 'Purchase_Frequency']].mean()
print(perfil)

# También puedes ver el conteo por cluster
print(df['Cluster'].value_counts())
```
