```python
import sqlite3
import pandas as pd

# Conectar a la base de datos (está en la misma carpeta)
conn = sqlite3.connect("clientes_segmentados.db")

# Consulta 1: promedio de valor de pedido por cluster
query1 = """
SELECT Cluster, 
       ROUND(AVG(Average_Order_Value), 2) AS valor_promedio
FROM clientes
GROUP BY Cluster
ORDER BY Cluster;
"""
print("=== Consulta 1: Valor promedio del pedido por cluster ===")
print(pd.read_sql_query(query1, conn))

# Consulta 2: cantidad de clientes por cluster
query2 = """
SELECT Cluster, 
       COUNT(*) AS total_clientes
FROM clientes
GROUP BY Cluster
ORDER BY Cluster;
"""
print("\n=== Consulta 2: Cantidad de clientes por cluster ===")
print(pd.read_sql_query(query2, conn))

conn.close()
```
